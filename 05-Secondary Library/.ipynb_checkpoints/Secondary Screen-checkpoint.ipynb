```python
read_counts.rename(columns={'Construct Barcode': 'sgRNA Sequence', 
 'Construct IDs': 'Target Gene Symbol', 
 'HCT116_RDB181_RepA': "dCas9-Kox1 RepA",
 'HCT116_RDB333_RepA': "dCas9-Kox1-MeCP2 RepA", 
 'HCT116_RDB181_RepB': "dCas9-Kox1 RepB", 
 'HCT116_RDB182_RepA': "dCas9-Zim3 RepA",
 'HCT116_RDB182_RepB': "dCas9-Zim3 RepB", 
 'HCT116_RDB333_RepB': "dCas9-Kox1-MeCP2 RepB", 
 'HCT116_RDB331_RepA': "Kox1-dCas9 RepA",
 'HCT116_RDB331_RepB': "Kox1-dCas9 RepB", 
 'HCT116_RDB332_RepA': "Zim3-dCas9 RepA", 
 "pDNA CP2084":'pDNA',
 "HCT116_RDB332_RepB":'Zim3-dCas9 RepB'}, inplace=True)
```

```python
read_counts.drop('Empty', axis=1, inplace=True)
```

```python
read_counts.columns
```

```python
columns_to_convert = ['dCas9-Kox1 RepA',
       'dCas9-Kox1-MeCP2 RepA', 'dCas9-Kox1 RepB', 'dCas9-Zim3 RepA',
       'dCas9-Zim3 RepB', 'dCas9-Kox1-MeCP2 RepB', 'Kox1-dCas9 RepA',
       'Kox1-dCas9 RepB', 'Zim3-dCas9 RepA', 'pDNA', 'Zim3-dCas9 RepB']
```

```python
# Check DataFrame shape
print(read_counts.shape)
```

```python
# Check for missing values
print(read_counts[columns_to_convert].isnull().sum())

for col in columns_to_convert:
    read_counts[col] = read_counts[col].astype(float)
```

```python
lognorms = pool.lognorm_columns(reads_df=read_counts, columns=columns_to_convert)
filtered_lognorms = pool.filter_pdna(lognorm_df=lognorms, pdna_cols=['pDNA'], z_low=-3)
print('Filtered ' + str(lognorms.shape[0] - filtered_lognorms.shape[0]) + ' rows due to low pDNA abundance')
```

```python
lfc_df = pool.calculate_lfcs(lognorm_df=filtered_lognorms, ref_col='pDNA', 
                             target_cols=['dCas9-Kox1 RepA',
       'dCas9-Kox1-MeCP2 RepA', 'dCas9-Kox1 RepB', 'dCas9-Zim3 RepA',
       'dCas9-Zim3 RepB', 'dCas9-Kox1-MeCP2 RepB', 'Kox1-dCas9 RepA',
       'Kox1-dCas9 RepB', 'Zim3-dCas9 RepA', 'Zim3-dCas9 RepB'])

```

```python
avg_replicate_lfc_df = lfc_df.copy()
replicate_columns = {
    'dCas9-Kox1': ['dCas9-Kox1 RepA', 'dCas9-Kox1 RepB'],
    'dCas9-Kox1-MeCP2': ['dCas9-Kox1-MeCP2 RepA', 'dCas9-Kox1-MeCP2 RepB'],
    'dCas9-Zim3': ['dCas9-Zim3 RepA', 'dCas9-Zim3 RepB'],
    'Kox1-dCas9': ['Kox1-dCas9 RepA', 'Kox1-dCas9 RepB'],
    'Zim3-dCas9': ['Zim3-dCas9 RepA', 'Zim3-dCas9 RepB']
}

for group, cols in replicate_columns.items():
    avg_replicate_lfc_df[group] = avg_replicate_lfc_df[cols].mean(axis=1)  

    avg_replicate_lfc_df.drop(columns=cols, inplace=True)
```

```python
avg_replicate_lfc_df.columns
```

```python
clt_avg_replicate_lfc_df = avg_replicate_lfc_df[avg_replicate_lfc_df['Target Gene Symbol'].str.contains('ONE_INTERGENIC')]

means = clt_avg_replicate_lfc_df[['dCas9-Kox1', 'dCas9-Kox1-MeCP2',
       'dCas9-Zim3', 'Kox1-dCas9', 'Zim3-dCas9']].mean()
std_devs = clt_avg_replicate_lfc_df[['dCas9-Kox1', 'dCas9-Kox1-MeCP2',
       'dCas9-Zim3', 'Kox1-dCas9', 'Zim3-dCas9']].std()

# Perform Z-score calculation for the dCas9 columns in the original avg_replicate_lfc_df
for col in ['dCas9-Kox1', 'dCas9-Kox1-MeCP2',
       'dCas9-Zim3', 'Kox1-dCas9', 'Zim3-dCas9']:
    avg_replicate_lfc_df[f'{col} (Z-score)'] = (avg_replicate_lfc_df[col] - means[col]) / std_devs[col]
```

```python
z_score_columns = avg_replicate_lfc_df.columns[avg_replicate_lfc_df.columns.str.contains('Z-score')]

correlation_matrix = avg_replicate_lfc_df[z_score_columns].corr(method='pearson')

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
```

```python
# Select Z-score columns and gene names for hue
#z_score_columns = ['Z_dCas9-Kox1', 'Z_dCas9-Kox1-MeCP2', 'Z_dCas9-Zim3', 'Z_Kox1-dCas9', 'Z_Zim3-dCas9']
z_score_columns = ['dCas9-Kox1 (Z-score)', 'dCas9-Kox1-MeCP2 (Z-score)', 'dCas9-Zim3 (Z-score)']

# Create a pairplot with hue and scatterplot on the lower triangle
custom_palette = {'Essential': '#ff7f0e', 'Non-essential': '#1f77b4'}  # You can adjust colors here

pair_plot = sns.pairplot(selected_secondary_GeneAnnoZscore, hue='Category', 
                         vars=z_score_columns, kind='scatter', corner = True, palette=custom_palette )
plt.setp(pair_plot._legend.get_texts(), fontsize=14) 
pair_plot._legend.set_title('')  # Set the legend title to an empty string to remove it

# Compute Pearson correlation matrix
correlation_matrix = selected_secondary_GeneAnnoZscore[z_score_columns].corr()

for i, j in zip(*np.tril_indices_from(correlation_matrix, -1)):
    ax = pair_plot.axes[i, j]
    if ax is not None:
        # Calculate Pearson correlation for each pair of variables
        corr_value = correlation_matrix.iloc[i, j]
        
        # Add the correlation text in the format 'r = XX' in the top-left corner
        ax.annotate(f'r = {corr_value:.2f}', 
                    xy=(0.1, 0.9), xycoords='axes fraction',
                    ha='left', va='top', fontsize=12, color='black')

# Set same limits for x and y axis and add x=y line only in the lower triangle scatter plots
for i, j in zip(*np.tril_indices_from(pair_plot.axes, -1)):  # Lower triangle
    ax = pair_plot.axes[i, j]
    if ax is not None:
        # Get limits
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()
        lims = [min(xlim[0], ylim[0]), max(xlim[1], ylim[1])]
        lims = [-14,5]
        # Set equal limits
        ax.set_xlim(lims)
        ax.set_ylim(lims)
        
        # Add x = y line
        ax.plot(lims, lims, ls='--', c='black')
# plt.xticks(fontsize=16)  # Increase x-axis label font size
# plt.yticks(fontsize=16)  # Increase y-axis label font size
custom_ticks = [0, -5, -10]

for ax in pair_plot.axes.flatten():
    if ax is not None:  # Check if the axis is not None
        ax.set_xlabel(ax.get_xlabel(), fontsize=13)  # Increase x-axis label font size
        ax.set_ylabel(ax.get_ylabel(), fontsize=13)  # Increase y-axis label font size
        ax.tick_params(labelsize=13)                 # Increase tick label font size
        ax.set_xticks(custom_ticks)
        ax.set_yticks(custom_ticks)

# plt.legend(fontsize=12)  # Adjust the legend font size here

plt.savefig(figPath+"CterminusScatter.pdf", dpi=300, bbox_inches='tight')
```

```python
# Create a plot to test
sns.set_style("white")
sns.set_context("paper")
#plt.switch_backend('svg')
custom_palette = {'Essential': '#ff7f0e', 'Non-essential': '#1f77b4'}  # You can adjust colors here

# List of all columns to plot
columns = ['dCas9-Kox1 (Z-score)', 'dCas9-Kox1-MeCP2 (Z-score)',
       'dCas9-Zim3 (Z-score)', 'Kox1-dCas9 (Z-score)', 'Zim3-dCas9 (Z-score)']
version = '1'

# Create scatter plots for each pair
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        plt.figure(figsize=(4, 4))
        ax = sns.scatterplot(x=columns[i], y=columns[j], hue = 'Category', 
                             data=selected_secondary_GeneAnnoZscore, palette=custom_palette )
        sns.despine()
        # Calculate the Pearson correlation
        corr = selected_secondary_GeneAnnoZscore[[columns[i], columns[j]]].corr(method = 'pearson').iloc[0, 1]
        
        # Plot title with correlation
        plt.text(0.05, 0.95, f'r = {corr:.2f}', ha='left', va='top', transform=plt.gca().transAxes, fontsize=14)

        #plt.title(f'Scatter Plot of {columns[i]} vs {columns[j]} (Pearson Correlation: {corr:.2f})')
        plt.xlabel(columns[i], fontsize=16)
        plt.ylabel(columns[j], fontsize=16)
        # Save the figure to desktop
        figure_name = f'Scatter_{columns[i]}_vs_{columns[j]}_v{version}.pdf'#pdf
       
        plt.legend(loc='lower right', fontsize='large')
        plt.xticks(fontsize=16)
        plt.yticks(fontsize=16)
        custom_ticks = [5, 0, -5, -10]
        ax.set_xticks(custom_ticks)
        ax.set_yticks(custom_ticks)
        ax.set_xlim(-15, 5)
#        ax.set_xlim(-110, 20)
        ax.set_ylim(-15, 5)
#        ax.set_ylim(-110, 20)
        ax.set_aspect('equal')
        ax.plot([0,1],[0,1], transform=ax.transAxes,color = 'black')

        plt.savefig(figPath + figure_name, format='pdf', dpi=300, bbox_inches='tight', transparent=True, pad_inches=0.1)


        #plt.show()
```
