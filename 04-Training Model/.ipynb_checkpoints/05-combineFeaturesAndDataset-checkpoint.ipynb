```python
import pandas as pd
#from rs3.seq import predict_seq
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# import DHS, Chip and ATAC file and make sure they have the same annotation for chromosome and 
# make doubleZscore for Nunez and InHouse from smaller(better) and large(worse) to larger(better) and smaller(worse)
pathIn = '../../../../data/2024/Manuscript/CleanedScaledData/'
DHS = pd.read_csv(f'{pathIn}DHS/DHSwithScaledData.csv')
DHS['chromosome'] = DHS['chromosome'].replace('chrY','chr24').copy()
DHS['chromosome'] = DHS['chromosome'].replace('chrX','chr23').copy()
DHS.loc[(DHS.DataSet == 'Nunez') | (DHS.DataSet == 'InHouse'), 'doubleZscore'] *= -1

CHIP = pd.read_csv(f'{pathIn}CHIP/CHIPwithScaledData.csv')
CHIP['chromosome'] = CHIP['chromosome'].replace('chrY','chr24').copy()
CHIP['chromosome'] = CHIP['chromosome'].replace('chrX','chr23').copy()
CHIP.loc[(CHIP.DataSet == 'Nunez') | (CHIP.DataSet == 'InHouse'), 'doubleZscore'] *= -1

ATAC = pd.read_csv(f'{pathIn}ATAC/ATACwithScaledData.csv')
ATAC['chromosome'] = ATAC['chromosome'].replace('chrY','chr24').copy()
ATAC['chromosome'] = ATAC['chromosome'].replace('chrX','chr23').copy()
ATAC.loc[(ATAC.DataSet == 'Nunez') | (ATAC.DataSet == 'InHouse'), 'doubleZscore'] *= -1

Orientation = pd.read_csv(f'{pathIn}Orientation/OrientationScaledData.csv')
```

```python
# checking every unique sgRNA Context Sequence in same Domain and DataSet and Target Gene Symbol has 
# the number of copies determines by the number of alternative datasets
count_of_3 = (DHS[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 3).sum()

sum((DHS[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 3)) == count_of_3
```

```python
count_of_7 = (ATAC[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 7).sum()
sum((ATAC[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 7)) == count_of_7
```

```python
count_of_8 = (CHIP[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 8).sum()
sum((CHIP[['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet']].value_counts() == 8)) == count_of_8
```

```python
DHS
```

```python
# find the mean doubleZscore and sum overlap with peaks
DHS_agg = DHS.groupby(['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet'], as_index=False).agg({"doubleZscore": ["mean"], "overlap with peak": ["sum"]}).rename(columns = {"overlap with peak": "DHS overlap with peak"})

ATAC_agg = ATAC.groupby(['Target Gene Symbol', 'chromosome', 
             'sgRNA Context Sequence', 'Domain',
             'DataSet'], as_index=False).agg({"doubleZscore": ["mean"], "overlap with peak": ["sum"]}).rename(columns = {"overlap with peak": "ATAC overlap with peak"})
ATAC_agg.columns = ['_'.join(col).strip() if col[1] else col[0] for col in ATAC_agg.columns]
DHS_agg.columns = ['_'.join(col).strip() if col[1] else col[0] for col in DHS_agg.columns]
```

```python
DHS_agg
```

```python
# didnt merge on ('doubleZscore','mean') because there is some minor numerical x to the e-17 difference 
mergedATACandDHS = pd.merge(ATAC_agg, DHS_agg, on = ['Target Gene Symbol', 'chromosome', 
                                                     'sgRNA Context Sequence', 'Domain',
                                                     'DataSet'])
mergedATACandDHS[['Domain','DataSet']].value_counts()
```

## proceed with more processing

```python
# Pivot the DataFrame
pivoted_CHIP = CHIP.pivot_table(index=['Target Gene Symbol', 
                               'chromosome', 
                               'sgRNA Context Sequence', 
                               'Domain',
                               'DataSet'], 
                        columns='chipseq', 
                        values='overlap with peak',
                        aggfunc='sum',                        # Aggregation function to handle multiple values
                        fill_value=0                          # Fill missing values with 0
                       ).reset_index()

pivoted_CHIP['H3K4me3'] = pivoted_CHIP[['H3K4me3_A549','H3K4me3_HCT116','H3K4me3_K562']].sum(axis=1)
pivoted_CHIP['H3k27ac'] = pivoted_CHIP[['H3k27ac_A549','H3k27ac_HCT116','H3k27ac_K562']].sum(axis=1)
pivoted_CHIP = pivoted_CHIP[['Target Gene Symbol', 'chromosome', 'sgRNA Context Sequence', 'Domain',
       'DataSet', 'H3K4me3', 'H3k27ac']]
pivoted_CHIP[['Domain','DataSet']].value_counts()
```

```python
pivoted_CHIP.head()
```

```python
mergedATACandDHSandOrientandCHIP = pd.merge(mergedATACandDHSandOrient,pivoted_CHIP)
```

```python
# orginal cleaned data to get sgRNA 'Cut' Position and sgRNA 'Cut' Site TSS Offset

cleanedAllData_path = '../../../../data/2024/Manuscript/CleanedData/'
# DatasetCombineCleaned.csv is the first version cleaned. It used the orginal phenotype calcuation 
originalCleanedData = pd.read_csv(cleanedAllData_path + 'DatasetCombineCleaned_v3.csv', low_memory=False)
originalCleanedData['sgRNA Context Sequence'] = originalCleanedData['sgRNA Context Sequence'].apply(lambda x: x.upper())
originalCleanedData = originalCleanedData[['Target Gene Symbol', 'chromosome',
       'sgRNA Context Sequence', 'sgRNA \'Cut\' Position', 'Strand of Target',
       'TSS Position', 'sgRNA \'Cut\' Site TSS Offset', 'Domain']]
originalCleanedData['chromosome'] = originalCleanedData['chromosome'].replace('chrY','chr24')
originalCleanedData['chromosome'] = originalCleanedData['chromosome'].replace('chrX','chr23')
originalCleanedData.head()
```

```python
mergedDataFeatureAvgZZ = pd.merge(mergedATACandDHSandOrientandCHIP, originalCleanedData.drop_duplicates())
len(mergedDataFeatureAvgZZ)
```

```python
mergedDataFeatureAvgZZ.columns
```

```python
# find average doubleZscore for the same guide and same domain across different datasets 
mergedDataFeatureAvgZZ = pd.DataFrame(mergedDataFeatureAvgZZ.groupby(['Target Gene Symbol', 'sgRNA Context Sequence', 'Domain', 
       'Strand of Target', 'ModifiedGuideStrand', 'Orientation', 'chromosome',
        'ATAC overlap with peak_sum',
       'DHS overlap with peak_sum', 'H3K4me3', 'H3k27ac',
       'sgRNA \'Cut\' Position', 'TSS Position', 'sgRNA \'Cut\' Site TSS Offset'])['doubleZscore'].mean()).reset_index()
mergedDataFeatureAvgZZ.head()
```

```python

len(mergedDataFeatureAvgZZ)
```

```python
# Create intervals of 25 between -5000 and 5000
interval_25 = list(range(-5000, 5025, 25))


# Create labels based on the new cut_list
labels_5kb = [f'[{interval_25[i]}, {interval_25[i+1]})' for i in range(len(interval_25)-1)]

# print("cut_list:", interval_25)
# print("labels:", labels_5kb)
# Categorize values into intervals with right=False except the last one

mergedDataFeatureAvgZZ['TSS Bucket (-5kb to 5kb)'] = pd.cut(mergedDataFeatureAvgZZ['sgRNA \'Cut\' Site TSS Offset'], 
                                         interval_25[:-1], labels=labels_5kb[:-1], right=False)

mergedDataFeatureAvgZZ['TSS Bucket (-5kb to 5kb)'] = np.where(
    mergedDataFeatureAvgZZ['sgRNA \'Cut\' Site TSS Offset'] >= interval_25[-2],
    pd.cut(mergedDataFeatureAvgZZ['sgRNA \'Cut\' Site TSS Offset'], 
           bins=interval_25[-2:], labels=[labels_5kb[-1]], right=True),
    mergedDataFeatureAvgZZ['TSS Bucket (-5kb to 5kb)']
)
mergedDataFeatureAvgZZ.head()
```
