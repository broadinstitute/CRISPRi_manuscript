```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
correlationsGilbertInHouse = pd.read_csv('../Data/CleanedScaledData/Correlation/correlationsGilbertInHouse.csv')
correlationsNunezInHouse = pd.read_csv('../Data/CleanedScaledData/Correlation/correlationsNunezInHouse.csv')
```

```python

correlationsGilbertInHouse['Phenotype'] = correlationsGilbertInHouse['Phenotype'].str.replace('ScaledRho', 'ScaledPhenoScore')
```

```python
correlationsNunezInHouse
```

```python
correlations = pd.merge(correlationsNunezInHouse[['Phenotype', 'Transformation', 'Correlation']].rename(columns = {'Correlation':
                                                                                          'Nunez vs InHouse Correlation'}),
         correlationsGilbertInHouse[['Phenotype', 'Transformation', 'Correlation']].rename(columns = {'Correlation':
                                                                                          'Gilbert vs InHouse Correlation'}),
         on = ['Phenotype', 'Transformation'])
```

```python
correlations
```

```python
sns.scatterplot(data=correlations, x='Nunez vs InHouse Correlation', y='Gilbert vs InHouse Correlation', 
                hue="Transformation")
```

```python
correlations['Absolute Gilbert vs InHouse Correlation'] = abs(correlations['Gilbert vs InHouse Correlation'])
```

```python
correlations['avgCorr'] = correlations[['Nunez vs InHouse Correlation','Absolute Gilbert vs InHouse Correlation']].mean(axis=1)
correlations = correlations.sort_values('avgCorr',ascending = True)
```

```python
graph_pathout = '../Figure/Figure 3/Individual Figures/'
# Group by 'Transformation' and calculate the mean correlation
mean_correlations = correlations.groupby('Transformation')['avgCorr'].mean().reset_index()

# Sort the DataFrame by 'avgCorr' in ascending order
mean_correlations = mean_correlations.sort_values(by='avgCorr')

# Create the point plot with vertical lines
plt.figure(figsize=(10, 6))
sns.pointplot(data=mean_correlations, x='Transformation', y='avgCorr', markers='|', markersize=10, color='black')


# Customize the plot
plt.xlabel('', fontsize=14)
plt.ylabel('Average Absolute Correlation', fontsize=14)
plt.title('Average Absolute Correlation by Transformation', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.ylim(0, 0.6)
plt.savefig(graph_pathout + "Average Absolute Correlation by Transformation.pdf", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
